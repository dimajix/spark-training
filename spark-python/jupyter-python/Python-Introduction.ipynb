# Conditions

```python
def print_birthday(name):
    if name in birthdays:
        bday = birthdays[name]
    else:
        bday = "<unknown>"
        
    print("%s is born on %s" % (name, bday))
    
        
print_birthday("Alice")
```

```python
print_birthday('Bob')
```

## Alternative Implementation

```python
def print_birthday(name):
    bday = birthdays.get(name, "<unknown>")
        
    print("%s is born on %s" % (name, bday))
    
print_birthday("Alice")
print_birthday("Bob")
```

## Inheritance

```python
class Shape:
    def get_area(self):
        return 0
    def get_circumference(self):
        return 0
    def show(self):
        print("Area=%f, Circumference=%f" %(self.get_area(), self.get_circumference()))
    
class Box(Shape):
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def get_area(self):
        return self.a*self.b
    def get_circumference(self):
        return 2*(self.a+self.b)
    
class Circle(Shape):
    def __init__(self,r):
        self.r = r
    def get_area(self):
        return 3.141*self.r*self.r
    def get_circumference(self):
        return 2*3.141*self.r
```

```python
box = Box(2,3)
box.show()

circle = Circle(4)
circle.show()
```

```python
isinstance(box,Box)
```

```python
isinstance(box,Shape)
```

```python
isinstance(box,Circle)
```

# Map & Filter

## Mapping

```python
def square(x):
    return x*x

s = map(square, range(1,10))
s
```

```python
[n for n in s]
```

```python
[n for n in s]
```

```python
for n in map(square, range(1,10)):
    print(n)
```

## Filtering

```python
def is_even(x):
    return x % 2 == 0

evens = filter(is_even, range(1,10))
evens
```

```python
[e for e in evens]
```

## Reducing

```python
from functools import reduce

def add(x,y):
    return x + y

reduce(add, range(1,10))
```

# Lambda Functions

```python
square = lambda x: x*x

s = map(square, range(1,10))
for n in s:
    print(n)
```

```python
for n in map(lambda x: x*x, range(1,10)):
    print(n)
```

# Generator Functions

```python
def generator(r):
    for x in r: 
        print("Evaluate for %d" % x)
        yield x*x
        
squares = generator(range(1,10))
squares
```

```python
for x in squares:
    print(x)
```
